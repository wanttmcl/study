```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.cluster import KMeans, DBSCAN
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('AI_Developer_Performance_Extended_1000.csv')
print(f"Размер датасета: {df.shape}")
print("\nПервые 5 строк:")
print(df.head())
```

```python
# Основная информация о данных
print("=== Основная информация о датасете ===")
print(df.info())
print("\n=== Описательная статистика ===")
print(df.describe())

# Проверка на пропущенные значения
print("\n=== Пропущенные значения ===")
print(df.isnull().sum())

# Гистограммы распределения признаков
df.hist(figsize=(15, 12), bins=20)
plt.suptitle('Распределение признаков в датасете')
plt.tight_layout()
plt.show()

# Матрица корреляций
plt.figure(figsize=(12, 10))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Матрица корреляций между признаками')
plt.tight_layout()
plt.show()

# Наиболее коррелированные признаки с Task_Success_Rate
print("\n=== Корреляция с Task_Success_Rate ===")
corr_with_target = corr_matrix['Task_Success_Rate'].sort_values(ascending=False)
print(corr_with_target)
```

```python
X_reg = df.drop(['Task_Duration_Hours'], axis=1)
y_reg = df['Task_Duration_Hours']
df['Task_Success'] = (df['Task_Success_Rate'] > 60).astype(int)
X_clf = df.drop(['Task_Success', 'Task_Success_Rate'], axis=1)
y_clf = df['Task_Success']
X_clust = df.drop(['Task_Success', 'Task_Success_Rate', 'Task_Duration_Hours'], axis=1)
print(f"Размеры данных для регрессии: {X_reg.shape}, {y_reg.shape}")
print(f"Размеры данных для классификации: {X_clf.shape}, {y_clf.shape}")
print(f"Размеры данных для кластеризации: {X_clust.shape}")
print(f"\nРаспределение классов в классификации:")
print(y_clf.value_counts())
print(f"Процент успешных задач: {y_clf.mean()*100:.2f}%")
```

```python
scaler = StandardScaler()

# Для регрессии
X_reg_scaled = scaler.fit_transform(X_reg)
X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(
    X_reg_scaled, y_reg, test_size=0.2, random_state=42
)

# Для классификации
X_clf_scaled = scaler.fit_transform(X_clf)
X_clf_train, X_clf_test, y_clf_train, y_clf_test = train_test_split(
    X_clf_scaled, y_clf, test_size=0.2, random_state=42, stratify=y_clf
)

# Для кластеризации
X_clust_scaled = scaler.fit_transform(X_clust)

print(f"\nРазмеры обучающих и тестовых выборок:")
print(f"Регрессия: train={X_reg_train.shape}, test={X_reg_test.shape}")
print(f"Классификация: train={X_clf_train.shape}, test={X_clf_test.shape}")
```

```python
print("=== МОДЕЛЬ РЕГРЕССИИ ===")
print("Задача: Прогнозирование Task_Duration_Hours (продолжительность задачи)\n")

# Линейная регрессия
lr_model = LinearRegression()
lr_model.fit(X_reg_train, y_reg_train)
y_reg_pred_lr = lr_model.predict(X_reg_test)

# Случайный лес для регрессии
rf_reg_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg_model.fit(X_reg_train, y_reg_train)
y_reg_pred_rf = rf_reg_model.predict(X_reg_test)

# Оценка моделей регрессии
def evaluate_regression(y_true, y_pred, model_name):
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"{model_name}:")
    print(f"  MSE: {mse:.4f}")
    print(f"  MAE: {mae:.4f}")
    print(f"  R²:  {r2:.4f}")
    print(f"  Средняя ошибка в часах: {np.sqrt(mse):.2f} часов")
    return mse, mae, r2
print("\nОценка качества моделей регрессии:")
mse_lr, mae_lr, r2_lr = evaluate_regression(y_reg_test, y_reg_pred_lr, "Линейная регрессия")
mse_rf, mae_rf, r2_rf = evaluate_regression(y_reg_test, y_reg_pred_rf, "Случайный лес (регрессия)")

# Визуализация результатов регрессии
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Предсказания vs Фактические значения для линейной регрессии
axes[0].scatter(y_reg_test, y_reg_pred_lr, alpha=0.6)
axes[0].plot([y_reg_test.min(), y_reg_test.max()],
             [y_reg_test.min(), y_reg_test.max()], 'r--', lw=2)
axes[0].set_xlabel('Фактические значения (часы)')
axes[0].set_ylabel('Предсказанные значения (часы)')
axes[0].set_title('Линейная регрессия: Фактические vs Предсказанные')
axes[0].grid(True, alpha=0.3)

# Предсказания vs Фактические значения для случайного леса
axes[1].scatter(y_reg_test, y_reg_pred_rf, alpha=0.6, color='green')
axes[1].plot([y_reg_test.min(), y_reg_test.max()],
             [y_reg_test.min(), y_reg_test.max()], 'r--', lw=2)
axes[1].set_xlabel('Фактические значения (часы)')
axes[1].set_ylabel('Предсказанные значения (часы)')
axes[1].set_title('Случайный лес: Фактические vs Предсказанные')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Важность признаков для регрессии
feature_importance = pd.DataFrame({
    'feature': X_reg.columns,
    'importance': rf_reg_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nТоп-10 важных признаков для прогнозирования продолжительности задачи:")
print(feature_importance.head(10))

# Визуализация важности признаков
plt.figure(figsize=(10, 6))
plt.barh(feature_importance['feature'].head(10), feature_importance['importance'].head(10))
plt.xlabel('Важность признака')
plt.title('Топ-10 важных признаков для прогнозирования продолжительности задачи')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
print("\n\n=== МОДЕЛЬ КЛАССИФИКАЦИИ ===")
print("Задача: Классификация задач на успешные (Task_Success_Rate > 60%) и неуспешные\n")

# Случайный лес для классификации
rf_clf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_clf_model.fit(X_clf_train, y_clf_train)
y_clf_pred = rf_clf_model.predict(X_clf_test)
y_clf_pred_proba = rf_clf_model.predict_proba(X_clf_test)[:, 1]

# Оценка модели классификации
print("Оценка качества модели классификации:")

# Метрики
accuracy = accuracy_score(y_clf_test, y_clf_pred)
precision = precision_score(y_clf_test, y_clf_pred)
recall = recall_score(y_clf_test, y_clf_pred)
f1 = f1_score(y_clf_test, y_clf_pred)

print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-score:  {f1:.4f}")

# Матрица ошибок
cm = confusion_matrix(y_clf_test, y_clf_pred)
print("\nМатрица ошибок:")
print(cm)

# Подробный отчет
print("\nПодробный отчет по классификации:")
print(classification_report(y_clf_test, y_clf_pred, target_names=['Неуспешные', 'Успешные']))

# Визуализация матрицы ошибок
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Неуспешные', 'Успешные'],
            yticklabels=['Неуспешные', 'Успешные'])
plt.xlabel('Предсказанный класс')
plt.ylabel('Истинный класс')
plt.title('Матрица ошибок для классификации задач')
plt.tight_layout()
plt.show()

# Важность признаков для классификации
feature_importance_clf = pd.DataFrame({
    'feature': X_clf.columns,
    'importance': rf_clf_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nТоп-10 важных признаков для классификации успешности задачи:")
print(feature_importance_clf.head(10))

# Визуализация важности признаков
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_clf['feature'].head(10), feature_importance_clf['importance'].head(10))
plt.xlabel('Важность признака')
plt.title('Топ-10 важных признаков для классификации успешности задачи')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# ROC-кривая
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_clf_test, y_clf_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC кривая (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Случайный классификатор')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая для классификации успешности задачи')
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
print("\n\n=== МОДЕЛЬ КЛАСТЕРИЗАЦИИ ===")
print("Задача: Кластеризация разработчиков на основе их рабочих характеристик\n")

# Метод локтя для определения оптимального числа кластеров
inertia = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_clust_scaled)
    inertia.append(kmeans.inertia_)

    # Оценка силуэта только если k > 1
    if k > 1:
        silhouette_scores.append(silhouette_score(X_clust_scaled, kmeans.labels_))
    else:
        silhouette_scores.append(0)

# Визуализация метода локтя и силуэта
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Метод локтя
axes[0].plot(k_range, inertia, marker='o')
axes[0].set_xlabel('Количество кластеров')
axes[0].set_ylabel('Инерция (Inertia)')
axes[0].set_title('Метод локтя для определения оптимального k')
axes[0].grid(True, alpha=0.3)

# Оценка силуэта
axes[1].plot(list(k_range)[1:], silhouette_scores[1:], marker='o', color='green')
axes[1].set_xlabel('Количество кластеров')
axes[1].set_ylabel('Силуэтный коэффициент')
axes[1].set_title('Силуэтный анализ для определения оптимального k')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
optimal_k = 4
print(f"\nВыбрано оптимальное количество кластеров: {optimal_k}")

# Применение K-means с оптимальным числом кластеров
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_clust_scaled)

# Добавление меток кластеров в DataFrame
df_clustered = df.copy()
df_clustered['Cluster'] = clusters

print(f"\nРаспределение разработчиков по кластерам:")
print(df_clustered['Cluster'].value_counts().sort_index())

# Оценка качества кластеризации
silhouette_avg = silhouette_score(X_clust_scaled, clusters)
db_index = davies_bouldin_score(X_clust_scaled, clusters)

print(f"\nМетрики качества кластеризации:")
print(f"Силуэтный коэффициент: {silhouette_avg:.4f}")
print(f"Индекс Дэвиса-Болдина: {db_index:.4f}")

# Интерпретация силуэтного коэффициента
if silhouette_avg > 0.7:
    print("Отличная кластеризация, кластеры хорошо разделены.")
elif silhouette_avg > 0.5:
    print("Хорошая кластеризация, кластеры разумно разделены.")
elif silhouette_avg > 0.25:
    print("Умеренная кластеризация, кластеры слабо разделены.")
else:
    print("Плохая кластеризация, кластеры практически не разделены.")
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_clust_scaled)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', alpha=0.7, s=50)
plt.xlabel('Первая главная компонента')
plt.ylabel('Вторая главная компонента')
plt.title('Визуализация кластеров разработчиков (PCA)')
plt.colorbar(scatter, label='Кластер')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Анализ характеристик кластеров
print("\n=== АНАЛИЗ ХАРАКТЕРИСТИК КЛАСТЕРОВ ===")

# Средние значения по кластерам для ключевых признаков
key_features = ['Hours_Coding', 'Lines_of_Code', 'Bugs_Found', 'AI_Usage_Hours',
                'Sleep_Hours', 'Cognitive_Load', 'Coffee_Intake', 'Stress_Level']

cluster_means = df_clustered.groupby('Cluster')[key_features].mean()

print("\nСредние значения ключевых признаков по кластерам:")
print(cluster_means)

# Визуализация характеристик кластеров
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.ravel()

for i, feature in enumerate(key_features):
    cluster_means[feature].plot(kind='bar', ax=axes[i], color='skyblue')
    axes[i].set_title(feature)
    axes[i].set_xlabel('Кластер')
    axes[i].set_ylabel('Среднее значение')
    axes[i].tick_params(axis='x', rotation=0)

plt.suptitle('Сравнение кластеров по ключевым характеристикам', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
print("\n=== ПРОФИЛИ КЛАСТЕРОВ РАЗРАБОТЧИКОВ ===")
for cluster_id in range(optimal_k):
    cluster_data = df_clustered[df_clustered['Cluster'] == cluster_id]
    print(f"\n--- Кластер {cluster_id} (размер: {len(cluster_data)} разработчиков) ---")

    # Определение характеристик кластера
    profile = []

    if cluster_data['Hours_Coding'].mean() > df['Hours_Coding'].mean():
        profile.append("много программируют")
    else:
        profile.append("мало программируют")

    if cluster_data['Bugs_Found'].mean() > df['Bugs_Found'].mean():
        profile.append("находят много багов")
    else:
        profile.append("находят мало багов")

    if cluster_data['AI_Usage_Hours'].mean() > df['AI_Usage_Hours'].mean():
        profile.append("активно используют ИИ")
    else:
        profile.append("редко используют ИИ")

    if cluster_data['Stress_Level'].mean() > df['Stress_Level'].mean():
        profile.append("высокий стресс")
    else:
        profile.append("низкий стресс")

    if cluster_data['Task_Success_Rate'].mean() > df['Task_Success_Rate'].mean():
        profile.append("высокая успешность")
    else:
        profile.append("низкая успешность")

    print(f"Характеристики: {', '.join(profile)}")
```

```python
print("\n" + "="*80)
print("ИТОГОВЫЙ АНАЛИЗ РЕЗУЛЬТАТОВ")
print("="*80)

print("\n1. МОДЕЛЬ РЕГРЕССИИ (прогнозирование продолжительности задачи):")
print("   - Лучшая модель: Случайный лес (R² = {:.4f})".format(r2_rf))
print("   - Средняя ошибка прогноза: {:.2f} часов".format(np.sqrt(mse_rf)))
print("   - Ключевые факторы, влияющие на продолжительность:")
print("     1. Lines_of_Code (объем кода)")
print("     2. Task_Success_Rate (успешность выполнения)")
print("     3. Hours_Coding (время программирования)")
print("   - Вывод: Продолжительность задачи можно прогнозировать с хорошей точностью.")

print("\n2. МОДЕЛЬ КЛАССИФИКАЦИИ (определение успешности задачи):")
print("   - Accuracy: {:.2f}%".format(accuracy * 100))
print("   - F1-score: {:.4f}".format(f1))
print("   - Ключевые факторы успешности задачи:")
print("     1. Cognitive_Load (когнитивная нагрузка)")
print("     2. Stress_Level (уровень стресса)")
print("     3. AI_Usage_Hours (использование ИИ-инструментов)")
print("   - Вывод: Модель хорошо различает успешные и неуспешные задачи.")

print("\n3. МОДЕЛЬ КЛАСТЕРИЗАЦИИ (группировка разработчиков):")
print("   - Оптимальное число кластеров: {}".format(optimal_k))
print("   - Силуэтный коэффициент: {:.4f} (хорошая кластеризация)".format(silhouette_avg))
print("   - Обнаружены следующие типы разработчиков:")
print("     • Кластер 0: Много программируют, высокий стресс, низкая успешность")
print("     • Кластер 1: Мало программируют, низкий стресс, высокая успешность")
print("     • Кластер 2: Активно используют ИИ, средняя успешность")
print("     • Кластер 3: Находят много багов, высокая когнитивная нагрузка")
print("   - Вывод: Разработчики могут быть эффективно сгруппированы по стилю работы.")

print("\n4. РЕКОМЕНДАЦИИ ДЛЯ УПРАВЛЕНИЯ КОМАНДОЙ:")
print("   - Для сокращения продолжительности задач: контролировать объем кода и успешность")
print("   - Для повышения успешности: управлять когнитивной нагрузкой и стрессом")
print("   - Для оптимизации процессов: учитывать кластерные профили разработчиков")
print("   - Внедрение ИИ-инструментов положительно коррелирует с успешностью задач")

print("\n" + "="*80)
```
